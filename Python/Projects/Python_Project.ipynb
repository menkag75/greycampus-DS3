```python

#import pandas library - for data manipulation and analysis.
import os, glob
import pandas as pd
import datetime as dt

# set the path
path = "C:/Users/asheesh/Documents/menka/Data Scientist/Python/Project"

#read the individual csv files and merge to a single .csv
all_files = glob.glob(os.path.join(path, "data_*.csv"))
df_from_each_file = (pd.read_csv(f, sep=',') for f in all_files)
df_merged   = pd.concat(df_from_each_file, ignore_index=True)
df_merged.to_csv( "merged_imus.csv")

# read  the merged csv and store it in data frame
my_df=pd.read_csv('merged_imus.csv')
#print output
my_df
```
